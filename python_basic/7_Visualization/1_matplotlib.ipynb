### 2. 그래프 구성요소

plot은 데이터를 가지고 그래프를 그린다.
matplotlib은 기본적으로 figure, axes, plot을 기본 구조로 한다.

```python
plt.figure()
plt.axes()
# X축은 지정하지 않으면서 자동으로 1씩 증가가 된다.
plt.plot([1,2,3,4])
plt.plot([3,4,5,6])
```

figure, axes를 생략해도 알아서 실행해 준다.

```python
plt.plot([1,2,3,4])
```

```python
plt.style.use('seaborn-pastel')

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())


# 이때 그래프에서 나타내는 눈금을 tick이라고 한다.
# 즉, 그래프의 x tick은 150이고 y tick은 5이다.
ax.set_xticks([0, 250, 500, 750, 1000])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

# 눈금을 문자로 하기 위해서는?
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation = 30, fontsize='small')

# 제목 입력하기
ax.set_title('random walk plot')

# 라벨 입력하기
ax.set_xlabel('Stages')
ax.set_ylabel('Values')

# 범례 표시하기
ax.legend(loc='best')
# loc는 범례가 위치할 곳을 의미한다. best를 주게 되면 현재 그래프에서 최적의 위치를 자동으로 찾는다.

ax.set_xticks([0, 250, 500, 750, 1000])
```

x축 y축 범위 수정하기

```python
ax.get_xlim()
# 현재 그래프의 x축 범위를 가져온다.
```

```python
# 이를 변경하려면,
ax.set_xlim([100,900])
```

```python
ax.set_ylim([-100,100])
```
