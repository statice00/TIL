- 데이터 출력해 확인

```python
data = response.json()['items']['item']

for i in range(len(data)):
    print('사고년도 : ' + data[i]['acc_year'])
    print('발생월일시 : ' + data[i]['occrrnc_dt'])
    print('발생위치 x좌표 : ' + data[i]['occrrnc_lc_x_crd'])
    print('발생위치 y좌표 : ' + data[i]['occrrnc_lc_y_crd'])
    print('발생위치 경도좌표 : ' + data[i]['lo_crd'])
    print('발생위치 위도좌표 : ' + data[i]['la_crd'])
    print('사망자수 : ', data[i]['dth_dnv_cnt'])
    print('부상자수 :', data[i]['injpsn_cnt'])
    print('\n')

    
totalCount = response.json()['totalCount']
print('total : ', totalCount)
```

- 데이터프레임 생성

```python
accident_tbl = pd.DataFrame( columns=('year', 'occur_date', 'occur_x', 'occur_y', '경도', '위도', '사망자수', '부상자수'))

for i in range(len(data)):
    accident_tbl.loc[len(accident_tbl)] = [data[i]['acc_year'],data[i]['occrrnc_dt'],data[i]['occrrnc_lc_x_crd'],data[i]['occrrnc_lc_y_crd'],data[i]['lo_crd'],data[i]['la_crd'],data[i]['dth_dnv_cnt'],data[i]['injpsn_cnt']]
```

- 시각화

```python
from folium.plugins import MarkerCluster
import folium
import base64

map = folium.Map(
    location=[37.482104027205, 126.95982056543], zoom_start=12.5)

occur_date = list(accident_tbl['occur_date'])
occur_x = list(accident_tbl['occur_x'])
occur_y = list(accident_tbl['occur_y'])
lat = list(accident_tbl['위도'])
lng = list(accident_tbl['경도'])
death = list(accident_tbl['사망자수'])
injury = list(accident_tbl['부상자수'])

#marker cluster로 표시
marker_cluster = MarkerCluster().add_to(map)

for i in range(len(occur_date)):
    tooltip =  f'● 발생년월시: {occur_date[i]} <br> ● 사망자수: {death[i]} <br> ● 부상자수: {injury[i]}'
    folium.Marker(
    location = [lat[i],lng[i]],
    tooltip= tooltip,
    icon=folium.Icon(color='Red',icon='map-marker')
     ).add_to(marker_cluster)


map
```
